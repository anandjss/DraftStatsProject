```python
import pandas as pd
import jupyter as jp
import numpy as np
import scipy.stats as st
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
# This is setting the variable "data" to the inputted data from the 'draftdata.csv' file

data= pd.read_csv("draftdata.csv")
print('hi')
data
```

```python
# Random lineplot graph just to make sure everything is working fine. Showing correlation between age when a player is drafted and # ppg averaged last year of college ball. Makes sense that if players are older (played more years in college), they would average more ppg in college their last year
sns.lineplot(x=data[" age"], y=data[" ppg"])
```

```python
# Once again, just a check to make sure everything is running smoothly
data.head(10)
```

```python
# Setting everything EXCEPT Nprod as independent variables, and setting Nprod as the dependent variable. Essentially, trying to see how all these different values can be used to predict a rookie in the NBA's total production (ppg + apg+ rbg)
from sklearn.model_selection import train_test_split
x=data.drop([' Nprod'], axis =1)
y=data[' Nprod']
```

```python
# splitting 70% of data for training, 30% for testing
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=0, test_size=0.3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# checking to make sure everything is running smoothly—There should be 35 rows here because 0.7 * 50
x_train
```

```python
# creating LinearRegression model and feeding data into it
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
y_Nprod_predicted=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
sc_lr = round(model.score(x_test, y_test) * 100, 2)
print("Accuracy: ", str(sc_lr) , " %")
```
